```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read in the data and
df = pd.read_csv('survey_results_public.csv')

df.shape

#Extracting the columns we need
df_new = df.iloc[:,np.r_[0,1,2,3,5,6,7,8,9,10,11,12,19:37,39:43,58]].reset_index(drop=True)

```

```python
# Handling missing data
df_new.dropna(subset =['DevType'],inplace=True)
df_new=df_new.reset_index(drop=True)
#df_new.dtypes
```

```python
'''The Column 'Devtype' contains multiple answers, The code in this cell obtains the unique answers '''

Unique_roles = set()
df_new['DevType']=df_new['DevType'].str.split(';')
for row in df_new['DevType']:
    for role in row:
        Unique_roles.add(role)
        
Unique_roles
```

```python
'''The Column 'Devtype' contains multiple answers, The code in this cell obtains the unique answers '''
df1 = df_new.dropna(subset =['LanguageHaveWorkedWith'])
Unique_roles = set()
df1['LanguageHaveWorkedWith'] = df1['LanguageHaveWorkedWith'].str.split(';')
for row in df1['LanguageHaveWorkedWith']:
     for role in row:
        Unique_roles.add(role)
        
Unique_roles
```

```python

def split_multiple_answers(df,column):
    '''
    INPUT 
        df - the column to be splitted
        column - the column name you want to obtain the unique answers from
        
    OUTPUT
    
        Unique_vals - A set containing all the possible (unique) answers in the column
    '''
    df1=df.dropna(subset = [column])
    Unique_vals = set()
    df1[column]=df1[column].str.split(';')
    for row in df1[column]:
        for role in row:
            Unique_vals.add(role)
    return(Unique_vals)

```

```python
#Calling the function to split the necessary columns

Unique_current_languages= split_multiple_answers(df_new,'LanguageHaveWorkedWith')
Unique_next_languages=split_multiple_answers(df_new,'LanguageWantToWorkWith')
Unique_current_db=split_multiple_answers(df_new,'DatabaseHaveWorkedWith')
Unique_next_db=split_multiple_answers(df_new,'DatabaseWantToWorkWith')
Unique_current_platform=split_multiple_answers(df_new,'PlatformHaveWorkedWith')
Unique_next_platform=split_multiple_answers(df_new,'PlatformWantToWorkWith')
Unique_current_Webframe=split_multiple_answers(df_new,'WebframeHaveWorkedWith')
Unique_next_Webframe=split_multiple_answers(df_new,'WebframeWantToWorkWith')
     
```

```python
def total_count(df, col1, col2, look_for):
    '''
    INPUT:
    df - the pandas dataframe you want to search
    col1 - the column name you want to look through
    col2 - the column you want to count values from
    look_for - a list of strings you want to search for in each row of df[col]

    OUTPUT:
    new_df - a dataframe of each look_for with the count of how often it shows up
    '''
    new_df = defaultdict(int)
    #loop through list of ed types
    for val in look_for:
        #loop through rows
        for idx in range(df.shape[0]):
            #if the ed type is in the row add 1
            if val in df[col1][idx]:
                new_df[val] += int(df[col2][idx])
    new_df = pd.DataFrame(pd.Series(new_df)).reset_index()
    new_df.columns = [col1, col2]
    new_df.sort_values('count', ascending=False, inplace=True)
    return new_df
```
